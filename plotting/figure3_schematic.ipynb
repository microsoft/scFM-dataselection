```python
import numpy as np
import matplotlib.pyplot as plt
```

# Set plotting parameters.

```python
linewidth = 4

plt.rcParams['axes.linewidth'] = linewidth
plt.rcParams.update({'font.size': 18})
```

# Define a curve representing how we might expect performance might be a function of pre-training dataset size.

```python
ymax = 0.9

x = np.linspace(1, 100, 100)
y = ymax - np.exp(-x * 0.06)

marked_point = 25
```

# Plot the curve, a horizontal line at the maximum metric observed, a horizontal line at 95% of the maximum metric observed, and a point representing the first dataset size surpassing 95% of the maximum metric observed.

```python
fig, ax = plt.subplots(1,1, figsize=(6.5, 6))

ax.plot(x,y, linewidth=linewidth)
ax.scatter(x=x[marked_point], y=y[marked_point], s=128)

ax.axhline(y=ymax, label="Best Performance", linestyle="--", linewidth=linewidth, color='red')
ax.axhline(y=0.65, label="Saturation Threshold", linestyle=":", linewidth=linewidth, color='red')

ax.annotate("Learning Saturation Point", xy=(x[marked_point] + 1, y[marked_point] - 0.01), xytext=(x[marked_point], 0.3), arrowprops=dict(arrowstyle="simple", color="black"))

ax.set_ylim(0,1)

ax.set_xlabel("Dataset Size")
ax.set_ylabel("Evaluation Metric")


# Remove axis ticks and only have X and Y axes
ax.spines[['right', 'top']].set_visible(False)
plt.xticks([])
plt.yticks([])

ax.legend()

plt.savefig('figures/figure3_schematic.png', transparent=True, dpi=300)
```
