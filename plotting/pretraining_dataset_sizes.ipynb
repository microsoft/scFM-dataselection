```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
random_sizes = [0.759, 7.4, 19, 37, 56, 75]
geometric_sketching_sizes = [0.953, 8.8, 23, 46, 65]
celltype_reweighted_sizes = [0.722, 7.1, 18, 34, 51]
```

```python
dataset_sizes_dict = {
    "downsampling_method": ["Random"] * 6 + ["Geometric Sketching"] * 5 + ["Celltype Reweighted"] * 5,
    "percentage": [1, 10, 25, 50, 75, 100] + [1, 10, 25, 50, 75] * 2,
    "sizes": random_sizes + geometric_sketching_sizes + celltype_reweighted_sizes
}
```

```python
df = pd.DataFrame.from_dict(dataset_sizes_dict)
df["downsampling_method"] = df.downsampling_method.astype('category')
df.downsampling_method.cat.set_categories(["Celltype Reweighted", "Geometric Sketching", "Random"], ordered=True)
```

```python
sns.barplot(df, y="sizes", x="percentage", hue="downsampling_method")

plt.xlabel("Percentage of Original Data")
plt.ylabel("Pretraining Dataset Size (G)")
plt.title("Pretraining Dataset Sizes by Downsampling Method")
plt.legend(title="Downsampling Method")

plt.savefig("dataset_sizes.png")
```
