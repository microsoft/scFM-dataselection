```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# todo remove Geneformer+tokenization
```

```python
inference_time_df = pd.read_csv("./metrics_csvs/timing_benchmark_results.csv")
inference_time_df = inference_time_df[inference_time_df.model != "Geneformer+tokenization"]
# Reorder the categories
inference_time_df['model'] = inference_time_df['model'].astype('category')
desired_order = ["Pre-trained PCA", "scVI", "SSL", "Geneformer", "SCimilarity"]
inference_time_df['model'] = inference_time_df['model'].cat.reorder_categories(desired_order, ordered=True)

```

```python
inference_time_df
```

```python
#tick_labels = ["Pre-trained PCA", "scVI", "SSL", "Geneformer", "SCimilarity"]

fig, ax = plt.subplots(1,2)
bar_linear = sns.barplot(inference_time_df,
                  x="model",
                  y="elapsed_time",
                  capsize=2,
                  ax = ax[0])

#bar_linear.set_xticklabels(tick_labels)

ax[0].set_xlabel("Model Architecture")
ax[0].set_ylabel("Elapsed Time (sec)")
ax[0].tick_params(axis='x', rotation=90)

bar_log = sns.barplot(inference_time_df,
                  x="model",
                  y="elapsed_time",
                  capsize=2,
                  ax = ax[1])

#bar_log.set_xticklabels(tick_labels)

ax[1].set_yscale('log')
ax[1].set_xlabel("Model Architecture")
ax[1].set_ylabel("Elapsed Time (sec)")
ax[1].tick_params(axis='x', rotation=90)

fig.set_size_inches(7, 4)
plt.tight_layout()


plt.savefig("timing_benchmark.png", bbox_inches="tight", dpi=1200)
```
