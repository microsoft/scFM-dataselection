```python
import numpy as np
import matplotlib.pyplot as plt
```

# Set plotting parameters.

```python
linewidth = 4

plt.rcParams['axes.linewidth'] = linewidth
plt.rcParams.update({'font.size': 18})
```

# Define two curves representing how we might expect performance might be a function of pre-training dataset size.

```python
x = np.linspace(1, 100, 100)
y = y = 0.8 - np.exp(-x * 0.06)
y2 = 0.9 - np.exp(-x * 0.06)
```

# Plot the two curves.

```python
fig, ax = plt.subplots(1,1, figsize=(6, 6))

ax.plot(x,y, label="Less Diverse", linewidth=linewidth)
ax.plot(x,y2, label="More Diverse", linewidth=linewidth)

# Set axis labels
ax.set_ylim(0,1)
ax.set_xlabel("Dataset Size")
ax.set_ylabel("Evaluation Metric")

# Remove axis ticks and only have X and Y axes
ax.spines[['right', 'top']].set_visible(False)
plt.xticks([])
plt.yticks([])


ax.legend()

ax.legend(title='Downsampling Method')

plt.savefig('figures/figure1_schematic.png', transparent=True, dpi=300)
```
